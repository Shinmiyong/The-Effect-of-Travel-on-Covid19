```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import seaborn as sns
```

```python
df = pd.read_csv('./tmap_total/tmap_total_1.csv')
```

```python
# multi_indexing
df.set_index(['년도','지역 시군구'], inplace= True)
```

```python
df.head(10)
```

```python
indexs = df.index
print(indexs)
```

```python
index = []
for i in range(0, len(indexs)):
    a = indexs[i][1]
    index.append(a)
print(index)
```

```python
# 팔레트 리스트로 설정
colors = []
for i in range(0, 10):
    if indexs[i][0] == 2020:
        colors.append('pink')
    else:
        colors.append('skyblue')
print(colors)
```

```python
font_path = '../malgun.ttf'
font_name = font_manager.FontProperties(fname = font_path).get_name()
rc('font', family = font_name)
```

```python
plt.figure(figsize= (16, 10))
#sns.barplot(data= df, x= '여행 횟수', y= index)
travel = sns.barplot(data= df, x= '여행 횟수', y= indexs, palette=colors)
travel.set_title('2019년/2020년 1월 여행 횟수', size = 20) 
travel.yaxis.set_visible(False)               # yaxis 안보이게 하기

ax = travel
for i, p in enumerate(ax.patches):
    width = p.get_width()
    plt.text(5+p.get_width(), p.get_y()+0.5*p.get_height(),
                 '{:1.0f}'.format(width),
                 ha='center', va='center', fontsize=20)
    
    plt.text(55, p.get_y()+0.5*p.get_height(), str(index[i]), ha='center', va='center', fontsize=20)

plt.show()
```
