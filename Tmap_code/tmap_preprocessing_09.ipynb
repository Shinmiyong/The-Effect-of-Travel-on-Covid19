```python
import pandas as pd
```

```python
df_2019= pd.read_csv('./tmap_prep_for_vis/tmap_19_1_monthly_count_top20.csv', index_col=0)
df_2020= pd.read_csv('./tmap_prep_for_vis/tmap_20_1_monthly_count_top20.csv', index_col=0)

df_2019.reset_index(inplace=True)
df_2020.reset_index(inplace=True)
```

```python
df_2019.head(10)
```

```python
a_list = []
for i in range(0, len(df_2019)):
    if df_2019['지역 시군구'][i] == '강원 정선군' or df_2019['지역 시군구'][i] == '경기 여주시' or df_2019['지역 시군구'][i] == '강원 고성군' or df_2019['지역 시군구'][i] == '경기 포천시' or df_2019['지역 시군구'][i] == '전북 무주군':
        data = df_2019.iloc[i]
        a_list.append(data)

df2_2019 = pd.DataFrame(a_list)
df2_2019.reset_index(drop= True, inplace= True)
df2_2019
```

```python
df_2020.head(10)
```

```python
a_list = []
for i in range(0, len(df_2019)):
    if df_2020['지역 시군구'][i] == '경기 여주시' or df_2020['지역 시군구'][i] == '강원 평창군' or df_2020['지역 시군구'][i] == '강원 고성군' or df_2020['지역 시군구'][i] == '제주 서귀포시' or df_2020['지역 시군구'][i] == '경기 포천시':
        data = df_2020.iloc[i]
        a_list.append(data)

df2_2020 = pd.DataFrame(a_list)
df2_2020.reset_index(drop= True, inplace= True)
df2_2020
```

```python
df2_2019['년도'] = '2019'
df2_2020['년도'] = '2020'

df3 = pd.concat([df2_2019, df2_2020], axis= 'rows', ignore_index= True)
df3 = df3[['년도', '지역 시군구', '여행 횟수']]
df3.sort_values(by= ['여행 횟수'], axis= 0, ascending= False, inplace= True)
df3.reset_index(drop= True, inplace= True)
df3.to_csv('./tmap_monthly_top5_ver2/tmap_monthly_top5_7_modified.csv')
```
